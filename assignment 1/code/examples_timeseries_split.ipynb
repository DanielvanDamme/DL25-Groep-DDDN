```python
from timeseries_split import (
    load_time_series,
    create_lagged_features,
    train_test_split_lagged,
    plot_lagged_chunks
)
```

```python
# Load and prepare
series = load_time_series('Xtrain.mat')
```

# Examples
Here follow some examples of how you can split the data into training and testing and visualising what that looks like. More extensive documentation can be found as comments above the function definition of 'train_test_split_lagged' in the file 'timeseries_split.py'.

```python
# One big split
n_lags = 5 # Number of steps looking back
lagged_df = create_lagged_features(series, n_lags)

# Parameters for splitting percentage
train_size = 0.8

# Get the chunks (train/test pairs)
splits = train_test_split_lagged(lagged_df, train_size)

plot_lagged_chunks(series, splits, n_lags=n_lags, stride = 0)
```

```python
# Full coverage different sized chunks no overlap
n_lags = 5
lagged_df = create_lagged_features(series=series, n_lags=n_lags, dropna=True)

train_size = 0.8
total_size = 200
stride = 200

splits = train_test_split_lagged(df=lagged_df, train_size=train_size, total_size=total_size, stride=stride)
plot_lagged_chunks(series=series, splits=splits, n_lags=n_lags, stride=stride)
```

```python
# Some overlap (difficutlt too see in graph, but maybe good for validation across time)
n_lags = 5
lagged_df = create_lagged_features(series=series, n_lags=n_lags, dropna=True)

train_size = 0.7
total_size = 300
stride = 100

splits = train_test_split_lagged(df=lagged_df, train_size=train_size, total_size=total_size, stride=stride)
plot_lagged_chunks(series=series, splits=splits, n_lags=n_lags, stride=stride)
```

```python
# Skipping some values in between chunks
n_lags = 5
lagged_df = create_lagged_features(series=series, n_lags=n_lags, dropna=True)

train_size = 0.7
total_size = 100
stride = 150

splits = train_test_split_lagged(df=lagged_df, train_size=train_size, total_size=total_size, stride=stride)
plot_lagged_chunks(series=series, splits=splits, n_lags=n_lags, stride=stride)
```
